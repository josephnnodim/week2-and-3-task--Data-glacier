# Checking for outliers in Transaction column using interquartile range

```python
#We could detect outliers in the Cab_data table using Interquartile range
#Any number less than the lower bounds or greater than the upper bounds is considered to be 
#an outlier, where
print("CHECKING FOR OUTLIERS")
print("\n")
print("Checking for outliers in the 'KM Travelled' column")
print("\n")

#We will do this in a few steps

#Step 1
print("step 1: Calculating the first and third quartile")

#Lets extract the column of interest from the Cab_data table

distance=Cab_data['KM Travelled']
distance

first_quartile, third_quartile= np.percentile(distance,[25,75])
print(first_quartile, third_quartile)

#Step 1
print("\n")
#Step 2
print("step 2: Calculating the Interquartile range")
IQR=third_quartile-first_quartile
print(IQR)


#Step 3
#Lets get the lower bound and upper bound using the
print("\n")
print("step 3: Calculating the lower and upper bounds")

lower_bound = first_quartile-(1.5 * IQR) 
upper_bound = third_quartile +(1.5 * IQR)

print("Lower bound")
print(lower_bound)
print("\n")
print("Upper bound")
print(upper_bound)
print("\n")

#Lets get the quartiles
#Let's get the interquartile range



#Step 4:
#Any number outside the range (-19.44 to 64.4) will be considered an outlier
print("Any number outside the range (-19.44 to 64.4) will be considered an outlier. Lets see the minimum and maximum values")
#We would check the minimum and maximum values to see if they fall within the range

print("\n")
print("Step 4: Minimum and maximum values")

print("Minimum distance")
print(min(distance))
print("\n")
print("Maximum distance")
print(max(distance))
```

### Since the minimum and maximum distance fall within the range (-19.44 to 64.4), there is no outlier in the "KM Travelled'  column of the Cab_data table

```python
import numpy as np
from statistics import mean
```

```python
outliers=[]

def detect_outliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```
